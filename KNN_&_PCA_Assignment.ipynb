# Theoretical

1. What is K-Nearest Neighbors (KNN) and how does it work?
   - K-Nearest Neighbors (KNN) is a simple yet powerful supervised machine learning algorithm used for both classification and regression. It works by finding the "K" closest data points to a new point, then using those neighbors to predict the new point's category (classification) or value (regression).

2. What is the difference between KNN Classification and KNN Regression?
   - KNN classification and KNN regression differ in their prediction objectives: classification predicts discrete categories, while regression predicts continuous numerical values. KNN classification assigns a data point to a class based on the majority class of its k-nearest neighbors, whereas KNN regression predicts a numerical value by averaging the target values of its k-nearest neighbors.

3. What is the role of the distance metric in KNN?
   - In K-Nearest Neighbors (KNN), the distance metric is crucial for determining the similarity between data points. It quantifies how close two data points are in the feature space, and this distance is used to find the "nearest" neighbors when classifying a new data point.

4. What is the Curse of Dimensionality in KNN?
   - The Curse of Dimensionality in k-NN refers to the phenomenon where the performance of the algorithm degrades as the number of features (dimensions) increases. In high-dimensional spaces, distances between data points become increasingly similar, making it harder for k-NN to effectively identify nearest neighbors and generalize well. This happens because the volume of space increases exponentially with each added dimension, making the data sparse and less informative for making predictions.

5. How can we choose the best value of K in KNN?
   - Choosing the optimal value of K in k-Nearest Neighbors (kNN) involves finding a balance between model performance and generalization. It's generally recommended to use odd numbers for K to avoid ties in classification, and cross-validation can help determine the best K for your specific dataset. The square root of the number of data points (n) can be used as a starting point, but it's important to experiment with different values and evaluate performance on a validation set.

6. What are KD Tree and Ball Tree in KNN?

The two algorithms KD-Tree & Ball Tree are used for finding the nearest neighbors of a query point. As mentioned in the name of the algorithms both the methods follow binary search tree method for finding nearest neighbors.

7. When should you use KD Tree vs. Ball Tree?
   - Slower than KD-Trees in low dimensions (dâ‰¤3) but a lot faster in high dimensions. Both are affected by the curse of dimensionality, but Ball-trees tend to still work if data exhibits local structure (e.g. lies on a low-dimensional manifold).

8. What are the disadvantages of KNN?
   - KNN (k-Nearest Neighbors) has several disadvantages, including high computational and memory costs, sensitivity to irrelevant features and noise, and struggles with high-dimensional data and imbalanced datasets.

9. How does feature scaling affect KNN?
   - Feature scaling significantly impacts the performance of k-Nearest Neighbors (KNN) algorithms. KNN relies on distance calculations to determine the nearest neighbors, and features with vastly different scales can disproportionately influence these distances, leading to skewed results. Scaling features ensures that all features contribute equally to the distance calculations, improving accuracy and robustness.

10. What is PCA (Principal Component Analysis)?
    - Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms a dataset with potentially correlated variables into a smaller set of uncorrelated principal components, preserving the most important information.

11. How does PCA work?
    - PCA (Principal Component Analysis) works by transforming a dataset into a new coordinate system, where the axes are orthogonal (perpendicular) principal components that capture the most variance in the data. This process reduces dimensionality while retaining the most important information, making it useful for data analysis, visualization, and other machine learning tasks.

12. What is the geometric intuition behind PCA?
    - PCA's geometric intuition lies in identifying directions (vectors) in a dataset where the data exhibits the most variability. It essentially finds an "axis" or "plane" (in 2D or 3D respectively) that best captures the spread of the data points. The first principal component (PC1) is the direction with the greatest variance, meaning the data points are most spread out along that line or plane. Subsequent principal components are orthogonal (at right angles) to the previous ones and capture the next most significant variance.

13. What is the difference between Feature Selection and Feature Extraction?
    - Feature selection chooses a subset of the most informative original features, while feature extraction creates new features by transforming the original data. Feature selection keeps the original data's characteristics, making it easier to interpret. Feature extraction can lead to new features that capture hidden patterns but may be harder to understand.

14. What are Eigenvalues and Eigenvectors in PCA?
    - In Principal Component Analysis (PCA), eigenvectors represent the directions (principal components) of maximum variance, while eigenvalues determine the magnitude of that variance along those directions.

15. How do you decide the number of components to keep in PCA?
    - To decide on the number of principal components (PCs) to keep in PCA, you should choose the smallest number that explains a sufficient amount of variance in your data, typically aiming for 95% or more. This can be achieved by plotting the cumulative explained variance ratio and selecting the point where the curve starts to flatten out, or by using the "elbow method" (scree plot) where the eigenvalues drop off significantly.

16. Can PCA be used for classification?
    - Yes, PCA can be used for classification, primarily as a preprocessing step or a feature transformation before feeding the data to a classifier. However, it's crucial to understand that PCA itself is a dimensionality reduction technique, not a classifier.

17. What are the limitations of PCA?
    - PCA (Principal Component Analysis) has limitations, including its assumption of linear relationships, sensitivity to outliers and scale, potential for information loss, and difficulty in interpreting the resulting components.

18. How do KNN and PCA complement each other?
    - PCA and KNN complement each other by enabling efficient use of KNN with high-dimensional data. PCA reduces dimensionality while preserving important variance, which can improve KNN's performance by simplifying calculations and potentially reducing overfitting. PCA can also help visualize high-dimensional data, making it easier to understand the data structure, according to IBM.

19. How does KNN handle missing values in a dataset?
    - Imputation with K-Nearest Neighbors (KNN) estimates missing values in a dataset by considering the values of the closest data points, determined by a distance metric like Euclidean distance. The missing value is then assigned the average of these nearest neighbors' values, weighted by their proximity.

20. What are the key differences between PCA and Linear Discriminant Analysis (LDA)?
    - PCA and LDA are both dimensionality reduction techniques, but they differ significantly in their goals and approach. PCA is unsupervised and aims to find a new set of orthogonal axes (principal components) that capture the maximum variance in the data. LDA, on the other hand, is supervised and focuses on maximizing the separation between different classes or groups. In essence, PCA is about capturing the most important features in the data, while LDA is about separating data points into distinct classes.

# 21. Train a KNN Classifier on the Iris dataset and print model accuracy.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the KNN classifier (you can adjust the number of neighbors 'n_neighbors')
knn = KNeighborsClassifier(n_neighbors=3)

# Train the KNN classifier on the training data
knn.fit(X_train, y_train)

# Make predictions on the test data
y_pred = knn.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

# Print the model accuracy
print(f"Model Accuracy: {accuracy:.4f}")
```

# 23. Train a KNN Classifier using different distance metrics (Euclidean and Manhattan) and compare accuracy.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the number of neighbors (k)
n_neighbors = 5

# Train and evaluate KNN with Euclidean distance
knn_euclidean = KNeighborsClassifier(n_neighbors=n_neighbors, metric='euclidean')
knn_euclidean.fit(X_train, y_train)
y_pred_euclidean = knn_euclidean.predict(X_test)
accuracy_euclidean = accuracy_score(y_test, y_pred_euclidean)
print(f"Accuracy with Euclidean distance (k={n_neighbors}): {accuracy_euclidean:.4f}")

# Train and evaluate KNN with Manhattan distance
knn_manhattan = KNeighborsClassifier(n_neighbors=n_neighbors, metric='manhattan')
knn_manhattan.fit(X_train, y_train)
y_pred_manhattan = knn_manhattan.predict(X_test)
accuracy_manhattan = accuracy_score(y_test, y_pred_manhattan)
print(f"Accuracy with Manhattan distance (k={n_neighbors}): {accuracy_manhattan:.4f}")

# Compare the accuracies
print("\nComparison of Accuracies:")
print(f"Euclidean Accuracy: {accuracy_euclidean:.4f}")
print(f"Manhattan Accuracy: {accuracy_manhattan:.4f}")

if accuracy_euclidean > accuracy_manhattan:
    print(f"\nEuclidean distance performed better with an accuracy of {accuracy_euclidean:.4f}.")
elif accuracy_manhattan > accuracy_euclidean:
    print(f"\nManhattan distance performed better with an accuracy of {accuracy_manhattan:.4f}.")
else:
    print("\nBoth Euclidean and Manhattan distances resulted in the same accuracy.")
```

# 24. Train a KNN Classifier with different values of K and visualize decision boundaried.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.inspection import DecisionBoundaryDisplay

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Using only the first two features for 2D visualization
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Values of K to experiment with
k_values = [1, 3, 5, 10]

# Create a figure with subplots for each K value
fig, axes = plt.subplots(1, len(k_values), figsize=(15, 5))

for i, k in enumerate(k_values):
    # Initialize and train the KNN Classifier
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # Plot the decision boundaries
    DecisionBoundaryDisplay.from_estimator(
        knn,
        X,
        cmap=plt.cm.RdYlBu,
        alpha=0.8,
        ax=axes[i],
        response_method="predict",
        plot_method="pcolormesh",
        shading="auto",
    )

    # Scatter plot of the training data points
    scatter = axes[i].scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=plt.cm.RdYlBu, edgecolor='k', s=20)

    # Set the title of the subplot
    axes[i].set_title(f'KNN (K={k})')
    axes[i].set_xlabel(iris.feature_names[0])
    axes[i].set_ylabel(iris.feature_names[1])

# Add a legend to the last subplot
handles, labels = scatter.legend_elements()
axes[-1].legend(handles, iris.target_names, loc="lower right", title="Classes")

plt.tight_layout()
plt.suptitle('Decision Boundaries of KNN Classifier with Different K Values', y=1.02)
plt.show()
```

# 25. Apply Feature Scaling before training a KNN model and compare results with unscaled data.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the number of neighbors (k)
n_neighbors = 5

# --- Train and evaluate KNN without feature scaling ---
knn_unscaled = KNeighborsClassifier(n_neighbors=n_neighbors)
knn_unscaled.fit(X_train, y_train)
y_pred_unscaled = knn_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)
print(f"Accuracy with unscaled data (k={n_neighbors}): {accuracy_unscaled:.4f}")

# --- Apply StandardScaler for feature scaling ---
scaler_standard = StandardScaler()
X_train_scaled_standard = scaler_standard.fit_transform(X_train)
X_test_scaled_standard = scaler_standard.transform(X_test)

# Train and evaluate KNN with StandardScaler
knn_scaled_standard = KNeighborsClassifier(n_neighbors=n_neighbors)
knn_scaled_standard.fit(X_train_scaled_standard, y_train)
y_pred_scaled_standard = knn_scaled_standard.predict(X_test_scaled_standard)
accuracy_scaled_standard = accuracy_score(y_test, y_pred_scaled_standard)
print(f"Accuracy with StandardScaler (k={n_neighbors}): {accuracy_scaled_standard:.4f}")

# --- Apply MinMaxScaler for feature scaling ---
scaler_minmax = MinMaxScaler()
X_train_scaled_minmax = scaler_minmax.fit_transform(X_train)
X_test_scaled_minmax = scaler_minmax.transform(X_test)

# Train and evaluate KNN with MinMaxScaler
knn_scaled_minmax = KNeighborsClassifier(n_neighbors=n_neighbors)
knn_scaled_minmax.fit(X_train_scaled_minmax, y_train)
y_pred_scaled_minmax = knn_scaled_minmax.predict(X_test_scaled_minmax)
accuracy_scaled_minmax = accuracy_score(y_test, y_pred_scaled_minmax)
print(f"Accuracy with MinMaxScaler (k={n_neighbors}): {accuracy_scaled_minmax:.4f}")

# Compare the results
print("\nComparison of Accuracies:")
print(f"Unscaled Accuracy:    {accuracy_unscaled:.4f}")
print(f"StandardScaler Accuracy: {accuracy_scaled_standard:.4f}")
print(f"MinMaxScaler Accuracy: {accuracy_scaled_minmax:.4f}")

if accuracy_scaled_standard > accuracy_unscaled and accuracy_scaled_standard >= accuracy_scaled_minmax:
    print("\nStandardScaler improved the model accuracy.")
elif accuracy_scaled_minmax > accuracy_unscaled and accuracy_scaled_minmax > accuracy_scaled_standard:
    print("\nMinMaxScaler improved the model accuracy.")
elif accuracy_scaled_standard == accuracy_scaled_minmax and accuracy_scaled_standard > accuracy_unscaled:
    print("\nBoth StandardScaler and MinMaxScaler improved the model accuracy.")
elif accuracy_scaled_standard > accuracy_unscaled:
    print("\nStandardScaler improved the model accuracy.")
elif accuracy_scaled_minmax > accuracy_unscaled:
    print("\nMinMaxScaler improved the model accuracy.")
elif accuracy_scaled_standard < accuracy_unscaled and accuracy_scaled_minmax < accuracy_unscaled:
    print("\nFeature scaling did not improve the model accuracy in this case.")
elif accuracy_scaled_standard == accuracy_unscaled and accuracy_scaled_minmax == accuracy_unscaled:
    print("\nFeature scaling did not change the model accuracy.")
elif accuracy_scaled_standard == accuracy_unscaled:
    print("\nStandardScaler did not change the model accuracy.")
elif accuracy_scaled_minmax == accuracy_unscaled:
    print("\nMinMaxScaler did not change the model accuracy.")
elif accuracy_scaled_standard == accuracy_scaled_minmax and accuracy_scaled_standard < accuracy_unscaled:
    print("\nBoth StandardScaler and MinMaxScaler decreased the model accuracy.")
elif accuracy_scaled_standard < accuracy_unscaled:
    print("\nStandardScaler decreased the model accuracy.")
elif accuracy_scaled_minmax < accuracy_unscaled:
    print("\nMinMaxScaler decreased the model accuracy.")
else:
    print("\nThe impact of feature scaling is not clearly discernible in this run.")
```

# 26. Train a PCA model on synthetic data and print the explained variance ratio for each component.

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Generate synthetic data with some correlation
np.random.seed(42)
n_samples = 100
n_features = 5

# Create a covariance matrix to introduce correlation
covariance_matrix = np.array([
    [1.0, 0.8, 0.5, 0.2, 0.1],
    [0.8, 1.0, 0.6, 0.3, 0.2],
    [0.5, 0.6, 1.0, 0.4, 0.3],
    [0.2, 0.3, 0.4, 1.0, 0.7],
    [0.1, 0.2, 0.3, 0.7, 1.0]
])

# Generate data using the covariance matrix
mean = np.zeros(n_features)
X = np.random.multivariate_normal(mean, covariance_matrix, size=n_samples)

# Apply PCA
pca = PCA()  # By default, it will keep all components
pca.fit(X)

# Get the explained variance ratio for each component
explained_variance_ratio = pca.explained_variance_ratio_

# Print the explained variance ratio
print("Explained Variance Ratio for Each Component:")
for i, ratio in enumerate(explained_variance_ratio):
    print(f"Component {i+1}: {ratio:.4f}")

# Optional: Visualize the explained variance ratio
plt.figure(figsize=(8, 6))
plt.bar(range(1, n_features + 1), explained_variance_ratio, alpha=0.8, align='center',
        label='Individual explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component')
plt.title('Explained Variance Ratio per Principal Component')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Optional: Calculate and print the cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
print("\nCumulative Explained Variance:")
for i, cum_var in enumerate(cumulative_explained_variance):
    print(f"Component {i+1}: {cum_var:.4f}")

# Optional: Visualize the cumulative explained variance
plt.figure(figsize=(8, 6))
plt.plot(range(1, n_features + 1), cumulative_explained_variance, marker='o', linestyle='-',
         label='Cumulative explained variance')
plt.xlabel('Number of principal components')
plt.ylabel('Cumulative explained variance')
plt.title('Cumulative Explained Variance')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 27. Apply PCA before training a KNN Classifier and compare accuracy with and without PCA.

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the number of neighbors (k)
n_neighbors = 5

# --- Train and evaluate KNN without PCA ---
knn_no_pca = KNeighborsClassifier(n_neighbors=n_neighbors)
knn_no_pca.fit(X_train, y_train)
y_pred_no_pca = knn_no_pca.predict(X_test)
accuracy_no_pca = accuracy_score(y_test, y_pred_no_pca)
print(f"Accuracy without PCA (k={n_neighbors}): {accuracy_no_pca:.4f}")

# --- Apply PCA before training KNN ---
# Standardize the data before applying PCA
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Choose the number of principal components to keep
n_components = 2  # You can experiment with different values

# Apply PCA
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Train and evaluate KNN with PCA
knn_pca = KNeighborsClassifier(n_neighbors=n_neighbors)
knn_pca.fit(X_train_pca, y_train)
y_pred_pca = knn_pca.predict(X_test_pca)
accuracy_pca = accuracy_score(y_test, y_pred_pca)
print(f"Accuracy with PCA (n_components={n_components}, k={n_neighbors}): {accuracy_pca:.4f}")

# Compare the results
print("\nComparison of Accuracies:")
print(f"Accuracy without PCA: {accuracy_no_pca:.4f}")
print(f"Accuracy with PCA:    {accuracy_pca:.4f}")

if accuracy_pca > accuracy_no_pca:
    print(f"\nPCA improved the model accuracy by {accuracy_pca - accuracy_no_pca:.4f}.")
elif accuracy_pca < accuracy_no_pca:
    print(f"\nPCA decreased the model accuracy by {accuracy_no_pca - accuracy_pca:.4f}.")
else:
    print("\nPCA did not change the model accuracy.")

# Optional: Print the explained variance ratio of the selected components
print("\nExplained Variance Ratio of the Selected Principal Components:")
print(pca.explained_variance_ratio_)
print(f"Total explained variance: {np.sum(pca.explained_variance_ratio_):.4f}")
```

# 28. Perform Hyperparameter Tuning on a KNN Classifier using GridSearchCV.


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature Scaling (important for KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the parameter grid to search
param_grid = {
    'n_neighbors': list(range(1, 31)),  # Test K values from 1 to 30
    'weights': ['uniform', 'distance'],  # Test both uniform and distance weighting
    'p': [1, 2]                           # Test both Manhattan (p=1) and Euclidean (p=2) distance
}

# Initialize the KNN Classifier
knn = KNeighborsClassifier()

# Initialize GridSearchCV
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
# - estimator: The model to tune (KNN classifier)
# - param_grid: The dictionary of parameters to search over
# - cv: The number of cross-validation folds (here, 5-fold cross-validation)
# - scoring: The metric to evaluate the model on (accuracy for classification)
# - verbose: Controls the verbosity of the output during the search
# - n_jobs: Number of CPU cores to use (-1 means use all available cores)

# Perform the hyperparameter search
grid_search.fit(X_train_scaled, y_train)

# Print the best parameters and the best score
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)

# Evaluate the best model on the test set
best_knn = grid_search.best_estimator_
y_pred_best = best_knn.predict(X_test_scaled)
test_accuracy_best = accuracy_score(y_test, y_pred_best)
print("Test Accuracy of the Best Model:", test_accuracy_best)
```

# 29. Train a KNN Classifier and check the number of misclassified samples.

# 32. Train a KNN Regressor and analyze the effect of different K values on performance.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Generate a synthetic dataset
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + 0.5 * np.random.randn(len(X))

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Range of K values to test
k_values = range(1, 21)
mse_scores = []

# Train and evaluate KNN Regressor for each K value
for k in k_values:
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    knn_regressor.fit(X_train, y_train)
    y_pred = knn_regressor.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

# Plot the MSE scores against K values
plt.figure(figsize=(10, 6))
plt.plot(k_values, mse_scores, marker='o', linestyle='-', color='b')
plt.title('Effect of K on KNN Regressor Performance (MSE)')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Mean Squared Error (MSE)')
plt.xticks(k_values)
plt.grid(True)
plt.show()

# Find the K value with the minimum MSE
best_k = k_values[np.argmin(mse_scores)]
min_mse = np.min(mse_scores)
print(f"\nOptimal K value: {best_k}")
print(f"Minimum Mean Squared Error: {min_mse:.4f}")

# Optional: Visualize the predictions for the best K value
knn_best = KNeighborsRegressor(n_neighbors=best_k)
knn_best.fit(X_train, y_train)
y_pred_best = knn_best.predict(X_test)

plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='darkorange', label='Actual Data')
plt.scatter(X_test, y_pred_best, color='cornflowerblue', label=f'Predicted Data (K={best_k})')
plt.xlabel('X')
plt.ylabel('y')
plt.title(f'KNN Regression with Optimal K={best_k}')
plt.legend()
plt.show()
```

# 35. Train a KNN Classifier using the KD Tree and Ball Tree algorithms and compare performance.

```python
import time
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the number of neighbors (k)
n_neighbors = 5

# --- KNN with KD Tree ---
start_time_kd = time.time()
knn_kd_tree = KNeighborsClassifier(n_neighbors=n_neighbors, algorithm='kd_tree')
knn_kd_tree.fit(X_train, y_train)
end_time_kd = time.time()
training_time_kd = end_time_kd - start_time_kd

start_time_predict_kd = time.time()
y_pred_kd = knn_kd_tree.predict(X_test)
end_time_predict_kd = time.time()
prediction_time_kd = end_time_predict_kd - start_time_predict_kd

accuracy_kd = accuracy_score(y_test, y_pred_kd)
print(f"KNN with KD Tree (k={n_neighbors}):")
print(f"  Training Time:   {training_time_kd:.4f} seconds")
print(f"  Prediction Time: {prediction_time_kd:.4f} seconds")
print(f"  Accuracy:        {accuracy_kd:.4f}")

# --- KNN with Ball Tree ---
start_time_ball = time.time()
knn_ball_tree = KNeighborsClassifier(n_neighbors=n_neighbors, algorithm='ball_tree')
knn_ball_tree.fit(X_train, y_train)
end_time_ball = time.time()
training_time_ball = end_time_ball - start_time_ball

start_time_predict_ball = time.time()
y_pred_ball = knn_ball_tree.predict(X_test)
end_time_predict_ball = time.time()
prediction_time_ball = end_time_predict_ball - start_time_predict_ball

accuracy_ball = accuracy_score(y_test, y_pred_ball)
print(f"\nKNN with Ball Tree (k={n_neighbors}):")
print(f"  Training Time:   {training_time_ball:.4f} seconds")
print(f"  Prediction Time: {prediction_time_ball:.4f} seconds")
print(f"  Accuracy:        {accuracy_ball:.4f}")

# Compare Performance
print("\nPerformance Comparison:")
print(f"Accuracy (KD Tree):   {accuracy_kd:.4f}")
print(f"Accuracy (Ball Tree): {accuracy_ball:.4f}")
print(f"Training Time (KD Tree):   {training_time_kd:.4f} seconds")
print(f"Training Time (Ball Tree): {training_time_ball:.4f} seconds")
print(f"Prediction Time (KD Tree):   {prediction_time_kd:.4f} seconds")
print(f"Prediction Time (Ball Tree): {prediction_time_ball:.4f} seconds")

if accuracy_kd > accuracy_ball:
    print("\nKD Tree achieved higher accuracy.")
elif accuracy_ball > accuracy_kd:
    print("\nBall Tree achieved higher accuracy.")
else:
    print("\nBoth algorithms achieved the same accuracy.")

if training_time_kd < training_time_ball:
    print("KD Tree trained faster.")
elif training_time_ball < training_time_kd:
    print("Ball Tree trained faster.")
else:
    print("Both algorithms had similar training times.")

if prediction_time_kd < prediction_time_ball:
    print("KD Tree predicted faster.")
elif prediction_time_ball < prediction_time_kd:
    print("Ball Tree predicted faster.")
else:
    print("Both algorithms had similar prediction times.")
```

# 38. Train a PCA model and analyze the effect of different numbers of components on accuracy.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Range of number of components to test
n_components_range = range(1, X_train_scaled.shape[1] + 1)
accuracy_scores = []

# Train and evaluate KNN for each number of components
for n_components in n_components_range:
    # Apply PCA
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    # Train KNN classifier
    knn = KNeighborsClassifier(n_neighbors=5)  # Using a fixed k for comparison
    knn.fit(X_train_pca, y_train)

    # Make predictions
    y_pred = knn.predict(X_test_pca)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Plot the accuracy against the number of components
plt.figure(figsize=(10, 6))
plt.plot(n_components_range, accuracy_scores, marker='o', linestyle='-', color='b')
plt.title('Effect of Number of PCA Components on KNN Accuracy')
plt.xlabel('Number of Principal Components')
plt.ylabel('Test Accuracy')
plt.xticks(n_components_range)
plt.grid(True)
plt.show()

# Find the optimal number of components based on accuracy
optimal_n_components = n_components_range[np.argmax(accuracy_scores)]
best_accuracy = np.max(accuracy_scores)
print(f"\nOptimal Number of Principal Components: {optimal_n_components}")
print(f"Best Accuracy with {optimal_n_components} Components: {best_accuracy:.4f}")

# Optionally, print the explained variance ratio for the optimal number of components
pca_optimal = PCA(n_components=optimal_n_components)
pca_optimal.fit(X_train_scaled)
print("\nExplained Variance Ratio for the Optimal Number of Components:")
print(pca_optimal.explained_variance_ratio_)
print(f"Total explained variance: {np.sum(pca_optimal.explained_variance_ratio_):.4f}")
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import time

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the number of neighbors (k) to use
n_neighbors = 5

# Different leaf_size values to test
leaf_sizes = [1, 5, 10, 20, 30, 50, 100]
accuracy_scores = {}
training_times = {}
prediction_times = {}

for leaf_size in leaf_sizes:
    # --- KNN with current leaf_size ---
    start_time_train = time.time()
    knn = KNeighborsClassifier(n_neighbors=n_neighbors, leaf_size=leaf_size)
    knn.fit(X_train, y_train)
    end_time_train = time.time()
    training_times[leaf_size] = end_time_train - start_time_train

    start_time_predict = time.time()
    y_pred = knn.predict(X_test)
    end_time_predict = time.time()
    prediction_times[leaf_size] = end_time_predict - start_time_predict

    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores[leaf_size] = accuracy

# Print the results
print(f"KNN Performance Comparison (k={n_neighbors}):")
print("--------------------------------------------------")
print("Leaf Size | Accuracy | Training Time (s) | Prediction Time (s)")
print("--------------------------------------------------")
for leaf_size in leaf_sizes:
    print(f"{leaf_size:<10}| {accuracy_scores[leaf_size]:<8.4f} | {training_times[leaf_size]:<17.4f} | {prediction_times[leaf_size]:<19.4f}")
print("--------------------------------------------------")

# Analyze the effect on accuracy
print("\nAnalysis of Accuracy:")
best_leaf_size_accuracy = max(accuracy_scores, key=accuracy_scores.get)
print(f"Best Accuracy ({accuracy_scores[best_leaf_size_accuracy]:.4f}) achieved with leaf_size = {best_leaf_size_accuracy}")

# Analyze the effect on training time
print("\nAnalysis of Training Time:")
best_leaf_size_train_time = min(training_times, key=training_times.get)
print(f"Shortest Training Time ({training_times[best_leaf_size_train_time]:.4f} s) achieved with leaf_size = {best_leaf_size_train_time}")

# Analyze the effect on prediction time
print("\nAnalysis of Prediction Time:")
best_leaf_size_predict_time = min(prediction_times, key=prediction_times.get)
print(f"Shortest Prediction Time ({prediction_times[best_leaf_size_predict_time]:.4f} s) achieved with leaf_size = {best_leaf_size_predict_time}")
```

# 42. Train a KNN Regressor and analyze the effect of different distance metrics on prediction error.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# Generate a synthetic dataset
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + 0.5 * np.random.randn(len(X))

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Number of neighbors (K) to use
n_neighbors = 5

# Distance metrics to test
distance_metrics = ['euclidean', 'manhattan', 'chebyshev', 'minkowski']
mse_scores = {}

# Train and evaluate KNN Regressor for each distance metric
for metric in distance_metrics:
    if metric == 'minkowski':
        # Minkowski requires the 'p' parameter
        knn_regressor = KNeighborsRegressor(n_neighbors=n_neighbors, metric=metric, p=3)  # Example p=3
        metric_name = 'minkowski (p=3)'
    else:
        knn_regressor = KNeighborsRegressor(n_neighbors=n_neighbors, metric=metric)
        metric_name = metric.capitalize()

    knn_regressor.fit(X_train, y_train)
    y_pred = knn_regressor.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_scores[metric_name] = mse

# Print the MSE scores for each distance metric
print(f"KNN Regressor Performance (k={n_neighbors}):")
for metric, mse in mse_scores.items():
    print(f"  {metric}: Mean Squared Error = {mse:.4f}")

# Visualize the predictions for each distance metric
plt.figure(figsize=(12, 8))
plt.scatter(X_test, y_test, color='darkorange', label='Actual Data')

for i, (metric, _) in enumerate(mse_scores.items()):
    if metric == 'minkowski (p=3)':
        knn_regressor = KNeighborsRegressor(n_neighbors=n_neighbors, metric='minkowski', p=3)
    else:
        knn_regressor = KNeighborsRegressor(n_neighbors=n_neighbors, metric=metric.lower())
    knn_regressor.fit(X_train, y_train)
    y_pred = knn_regressor.predict(X_test)
    plt.plot(X_test, y_pred, label=f'Predicted ({metric})')

plt.xlabel('X')
plt.ylabel('y')
plt.title('KNN Regression with Different Distance Metrics')
plt.legend()
plt.grid(True)
plt.show()

# Find the distance metric with the minimum MSE
best_metric = min(mse_scores, key=mse_scores.get)
min_mse = mse_scores[best_metric]
print(f"\nDistance Metric with Minimum MSE: {best_metric}")
print(f"Minimum Mean Squared Error: {min_mse:.4f}")
```

# 43. Train a KNN Classifier and evaluate using ROC-AUC score.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import label_binarize # import label_binarize from sklearn.preprocessing

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# For ROC-AUC, it's best to work with binary or multi-label data.
# Let's convert the Iris dataset into a binary classification problem
# by considering only two classes (e.g., Iris-setosa and Iris-versicolor)
X_binary = X[y != 2]
y_binary = y[y != 2]

# Split the binary dataset into training and testing sets
X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(
    X_binary, y_binary, test_size=0.3, random_state=42
)

# Initialize the KNN classifier
n_neighbors = 5
knn_binary = KNeighborsClassifier(n_neighbors=n_neighbors)

# Train the KNN classifier on the binary training data
knn_binary.fit(X_train_binary, y_train_binary)

# Make probability predictions on the binary test data
# predict_proba returns probabilities of the sample for each class
y_proba_binary = knn_binary.predict_proba(X_test_binary)[:, 1]

# Calculate the ROC-AUC score for the binary case
roc_auc_binary = roc_auc_score(y_test_binary, y_proba_binary)
print(f"ROC-AUC Score (Binary Classification, k={n_neighbors}): {roc_auc_binary:.4f}")

# Plot the ROC curve for the binary case
fpr_binary, tpr_binary, thresholds_binary = roc_curve(y_test_binary, y_proba_binary)

plt.figure(figsize=(8, 6))
plt.plot(fpr_binary, tpr_binary, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_binary:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Binary Classification')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# --- ROC-AUC for Multi-class (One-vs-Rest) ---
# For multi-class, ROC-AUC is typically calculated using a one-vs-rest approach
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Binarize the labels for multi-class ROC
y_binarized = label_binarize(y_test, classes=[0, 1, 2])
n_classes = y_binarized.shape[1]

knn_multi = KNeighborsClassifier(n_neighbors=n_neighbors)
knn_multi.fit(X_train, y_train)
y_proba_multi = knn_multi.predict_proba(X_test)

# Calculate ROC-AUC for each class using one-vs-rest
roc_auc_multi = {}
fpr_multi = {}
tpr_multi = {}

for i in range(n_classes):
    roc_auc_multi[i] = roc_auc_score(y_binarized[:, i], y_proba_multi[:, i])
    fpr_multi[i], tpr_multi[i], _ = roc_curve(y_binarized[:, i], y_proba_multi[:, i])

# Plot ROC curves for each class
plt.figure(figsize=(8, 6))
colors = ['blue', 'red', 'green']
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr_multi[i], tpr_multi[i], color=color, lw=2,
             label=f'ROC curve (class {i}, AUC = {roc_auc_multi[i]:.4f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) - Multi-class (One-vs-Rest)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Print the ROC-AUC scores for each class
print("\nROC-AUC Score (Multi-class, One-vs-Rest):")
for i in range(n_classes):
    # Access roc_auc_multi using the loop variable 'i' as a key
    # This ensures you are getting the correct ROC-AUC value for each class
    print(f"Class {i} ({iris.target_names[i]}): {roc_auc_multi.get(i, 0):.4f}")
```

# 47. Train a KNN Classifier and visualize the decision boundary.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import ListedColormap

# Load the Iris dataset (using only the first two features for 2D visualization)
iris = load_iris()
X = iris.data[:, :2]
y = iris.target
target_names = iris.target_names

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Initialize the KNN classifier
n_neighbors = 5
knn = KNeighborsClassifier(n_neighbors=n_neighbors)

# Train the KNN classifier
knn.fit(X_train, y_train)

# Define the range of the decision boundary plot
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict the class labels for all the points in the grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Create a color map
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# Plot the decision boundary
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)

# Plot the training points
for i, color in enumerate(cmap_bold.colors):
    idx = np.where(y_train == i)
    plt.scatter(X_train[idx, 0], X_train[idx, 1], c=color,
                label=target_names[i], cmap=cmap_bold, edgecolor='k', s=20)

plt.xlabel('Standardized Feature 1')
plt.ylabel('Standardized Feature 2')
plt.title(f'KNN Decision Boundary (k={n_neighbors})')
plt.legend()
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.grid(alpha=0.3)
plt.show()

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import ListedColormap
from sklearn.metrics import accuracy_score

# Load the Iris dataset (using only the first two features for 2D visualization)
iris = load_iris()
X = iris.data[:, :2]
y = iris.target
target_names = iris.target_names

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Initialize the KNN classifier
n_neighbors = 5
knn = KNeighborsClassifier(n_neighbors=n_neighbors)

# Train the KNN classifier
knn.fit(X_train, y_train)

# Define the range of the decision boundary plot
x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Predict the class labels for all the points in the grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Create a color map
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# Plot the decision boundary
plt.figure(figsize=(10, 8))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=cmap_light)

# Plot the training points
for i, color in enumerate(cmap_bold.colors):
    idx = np.where(y_train == i)
    plt.scatter(X_train[idx, 0], X_train[idx, 1], c=color,
                label=target_names[i], cmap=cmap_bold, edgecolor='k', s=20)

plt.xlabel('Standardized Feature 1')
plt.ylabel('Standardized Feature 2')
plt.title(f'KNN Decision Boundary (k={n_neighbors})')
plt.legend()
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.grid(alpha=0.3)
plt.show()

# Evaluate the accuracy on the test set
accuracy = accuracy_score(y_test, knn.predict(X_test))
print(f"\nTest Accuracy: {accuracy:.4f}")
```

# 48. Train a PCA model and analyze the effect of different numbers of components on data variance.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Standardize the data (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA with different numbers of components
n_components_range = range(1, X_scaled.shape[1] + 1)
explained_variance_ratios = []

for n_components in n_components_range:
    pca = PCA(n_components=n_components)
    pca.fit(X_scaled)
    explained_variance_ratios.append(pca.explained_variance_ratio_)

# Analyze the explained variance
plt.figure(figsize=(12, 6))

# Plot individual explained variance for each component
plt.subplot(1, 2, 1)
for i, ratios in enumerate(explained_variance_ratios):
    plt.plot(range(1, len(ratios) + 1), ratios, label=f'n_components={i+1}')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Individual Explained Variance Ratio vs. Component')
plt.legend()
plt.grid(True)

# Plot cumulative explained variance
plt.subplot(1, 2, 2)
cumulative_variance = [np.cumsum(ratios) for ratios in explained_variance_ratios]
for i, cum_var in enumerate(cumulative_variance):
    plt.plot(range(1, len(cum_var) + 1), cum_var, label=f'n_components={i+1}')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Explained Variance Ratio')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# --- Further Analysis ---

print("\nExplained Variance Ratio for Different Numbers of Components:")
for i, ratios in enumerate(explained_variance_ratios):
    print(f"\nNumber of Components: {i+1}")
    for j, ratio in enumerate(ratios):
        print(f"  Component {j+1}: {ratio:.4f} ({(ratio * 100):.2f}%)")

print("\nCumulative Explained Variance for Different Numbers of Components:")
for i, cum_var in enumerate(cumulative_variance):
    print(f"\nNumber of Components: {i+1}")
    for j, variance in enumerate(cum_var):
        print(f"  Up to Component {j+1}: {variance:.4f} ({(variance * 100):.2f}%)")

# --- Scree Plot (Alternative Visualization) ---
eigenvalues = []
for n_components in n_components_range:
    pca = PCA(n_components=n_components)
    pca.fit(X_scaled)
    eigenvalues.append(pca.explained_variance_)

plt.figure(figsize=(8, 6))
for i, evals in enumerate(eigenvalues):
    plt.plot(range(1, len(evals) + 1), evals, marker='o', linestyle='-', label=f'n_components={i+1}')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue')
plt.title('Scree Plot (Eigenvalues vs. Component Number)')
plt.legend()
plt.grid(True)
plt.show()
```
